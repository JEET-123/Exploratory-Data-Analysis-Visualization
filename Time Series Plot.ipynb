# `Time Series` Plot 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = [8, 4]
plt.rcParams['figure.dpi'] = 140
```

```python
df = pd.read_csv('daily-temperature.csv')
df
```

```python
df.info()
```

```python
df = pd.read_csv('daily-temperature.csv', parse_dates=True)
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df.set_index('Date', inplace=True)
```

```python
df.head()
```

# Line and Scatter Plot 

```python
df.plot(color = 'red')
plt.ylabel('Temp')
plt.title('Line Plot')
plt.show()
```

```python
df.plot(style = '.', color = 'red')
plt.ylabel('Temp')
plt.title('Line Plot')
plt.show()
```

## Subplots 

```python
groups = df.groupby(pd.Grouper(freq = 'A'))
```

```python
keys = groups.groups.keys()
```

```python
keys
```

```python
groups.get_group('1981-12-31')
```

```python
for key in keys:
    print(key)
```

```python
key
```

```python
groups.get_group(key)['Temp'].values
```

```python
years = pd.DataFrame()
for key in keys:
    years[key] = groups.get_group(key)['Temp'].values
```

```python
years
```

```python
years.plot(subplots = True, figsize = (10, 30))
plt.show()
plt.tight_layout()
```

# Heatmap
